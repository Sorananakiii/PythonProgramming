# Regular expressions

```python
# import regular expression operator
import re
```

```python
# Search for lines that contain 'From'

hand = open('data/mbox-short.txt')
for line in hand:
    line = line.rstrip()
    if re.search('From:', line):
        print(line)
```

```python
# Search for lines that start with 'From

hand = open('data/mbox-short.txt')
for line in hand:
    line = line.rstrip()
    if re.search('^From:', line):
        print(line)
```

```python
# Search for lines that start with 'F', followed by
# 2 characters, followed by 'm:'

hand = open('data/mbox-short.txt')
for line in hand:
    line = line.rstrip()
    if re.search('^F..m:', line):
        print(line)
```

```python
# Search for lines that start with From and have an at sign
hand = open('data/mbox-short.txt')
for line in hand:
    line = line.rstrip()
    if re.search('^From:.+@', line):
        print(line)
```

The findall() method searches the string in the second argument and returns a
list of all of the strings that look like email addresses. We are using a two-character
sequence that matches a non-whitespace character (\S).

```python
s = 'A message from csev@umich.edu to cwen@iupui.edu about meeting @2PM'
lst = re.findall('\S+@\S+', s)
print(lst)
```

```python
# Search for lines that have an at sign between characters

hand = open('data/mbox-short.txt')
for line in hand:
    line = line.rstrip()
    x = re.findall('\S+@\S+', line)
    if len(x) > 0:
        print(x)
```

```python
# Search for lines that have an at sign between characters
# The characters must be a letter or 

hand = open('data/mbox-short.txt')
for line in hand:
    line = line.rstrip()
    x = re.findall('[a-zA-Z0-9]\S+@\S+[a-zA-Z]', line)
    if len(x) > 0:
        print(x)
```

```python
# Search for lines that have an at sign between characters
# The characters must be a letter or number
# The results will be slightly more accurate than re07.py for email addresses

hand = open('data/mbox-short.txt')
for line in hand:
    line = line.rstrip()
    x = re.findall('[a-zA-Z0-9\-.]\S+@[a-zA-Z0-9].\S+[a-zA-Z]', line)
    if len(x) > 0:
        print(x)
```

```python
# Search for lines that start 'X' followed by any non whitespace
# characters and ':' then output the first group of non whitespace
# characters that follows

hand = open('data/mbox-short.txt')
for line in hand:
    line = line.rstrip()
    x = re.findall('^X\S*: (\S+)', line)
    if not x: continue
    print(x)
```

```python
# Search for lines that start with 'X' followed by any non
# whitespace characters and ':'
# followed by a space and any number.
# The number can include a decimal.

hand = open('data/mbox-short.txt')
for line in hand:
    line = line.rstrip()
    if re.search('^X\S*: [0-9.]+', line):
        print(line)
```

```python
# Search for lines that start with 'X' followed by any
# non whitespace characters and ':' followed by a space
# and any number. The number can include a decimal.
# Then print the number if it is greater than zero.

hand = open('data/mbox-short.txt')
for line in hand:
    line = line.rstrip()
    x = re.findall('^X\S*: ([0-9.]+)', line)
    if len(x) > 0:
        print(x)
```

```python
# Search for lines that start with 'Details: rev='
# followed by numbers and '.'
# Then print the number if it is greater than zero

hand = open('data/mbox-short.txt')
for line in hand:
    line = line.rstrip()
    x = re.findall('^Details:.*rev=([0-9.]+)', line)
    if len(x) > 0:
        print(x)
```

```python
# Search for lines that start with From and a character
# followed by a two digit number between 00 and 99 followed by ':'
# Then print the number if it is greater than zero

hand = open('data/mbox-short.txt')
for line in hand:
    line = line.rstrip()
    x = re.findall('^From .* ([0-9][0-9]):', line)
    if len(x) > 0: print(x)
```

```python
# Search for lines that contain 'Author:' followed by any characters,
# an at sign, and any non whitespace character
# Then print the character group that follows the at sign

hand = open('data/mbox-short.txt')
for line in hand:
    line = line.rstrip()
    x = re.findall('Author:.*@(\S+)', line)
    if not x: continue
    print(x)
```

```python
# Search for lines that contain 'New Revision: ' followed by a number
# Then turn the number into a float and append it to nums
# Finally print the length and the average of nums

fname = input('Enter file:')
hand = open(fname)
nums = list()
for line in hand:
    line = line.rstrip()
    x = re.findall('New Revision: ([0-9]+)', line)
    if len(x) == 1:
        val = float(x[0])
        nums.append(val)
print(len(nums))
print(int(sum(nums)/len(nums)))
```

# Summary

While this only scratched the surface of regular expressions, we have learned a bit
about the language of regular expressions. They are search strings with special
characters in them that communicate your wishes to the regular expression system
as to what defines matching and what is extracted from the matched strings.
Here are some of those special characters and character sequences:

```python
 Matches the beginning of the line.
$ Matches the end of the line.
. Matches any character (a wildcard).
\d: digit
\w: word character
\W: non-word character
\s: whitespace
* Applies to the immediately preceding character(s) and indicates to match zero or more times.
*? Applies to the immediately preceding character(s) and indicates to match zero or more times in non-greedy mode.
+ Applies to the immediately preceding character(s) and indicates to match one or more times.
+? Applies to the immediately preceding character(s) and indicates to match one or more times in non-greedy mode.
```

```python
? Applies to the immediately preceding character(s) and indicates to match zero or one time.
?? Applies to the immediately preceding character(s) and indicates to match zero or one time in non-greedy mode.
[aeiou] Matches a single character as long as that character is in the specified set.

In this example, it would match a, e, i, o, or u, but no other characters.

[a-z0-9] You can specify ranges of characters using the minus sign. This example
is a single character that must be a lowercase letter or a digit.

[A-Za-z] When the first character in the set notation is a caret, it inverts the
logic. This example matches a single character that is anything other than an
uppercase or lowercase letter.

( ) When parentheses are added to a regular expression, they are ignored for the
purpose of matching, but allow you to extract a particular subset of the matched
string rather than the whole string when using findall().

\b Matches the empty string, but only at the start or end of a word.
\B Matches the empty string, but not at the start or end of a word.
\d Matches any decimal digit; equivalent to the set [0-9].
\D Matches any non-digit character; equivalent to the set [0-9].
```

```python
pas = 'password1234'
# re.search(r'\w\w\w\w\w\w\w\w\d\d\d\d', pas)
re.findall(r'\w{8}\d{4}', pas)
```

```python
# Quantifiers : Once or more times (+)
t = 'Date of start : 4-3, Date of registration : 10-04.'
re.findall(r'\d+-\d+', t)
```

```python
# Quantifiers : zero or more times (*)
t2 = 'The concert was amazing! @ameki!a @joh&&n @mary90'
re.findall(r'@\w+\W*\w+', t2)
```

```python
# Quantifiers : zero or once times (?)
t3 = 'The color of this image is amazing. However, the colour blue could be brighter'
re.findall(r'colou?r', t3) # immediately to the left : is this case ? applies to u not to colou
```

```python
# Quantifiers : n times at least and m times at most ({n, m})
t4 = 'John: 1-966-847-3131 Michelle: 54-908-42-42424'
re.findall(r'\d{1,2}-\d{3}-\d{2,3}-\d{4,}', t4)
```

```python
#  Matches the beginning of the line
t5 = 'the 80s music was much better that the 90s'
print(re.findall(r'the\s\d+s', t5))
print(re.findall(r'^the\s\d+s', t5))
```

```python
# $ Matches the end of the line
t6 = 'the 80s music was much better that the 90s'
print(re.findall(r'the\s\d+s', t6))
print(re.findall(r'the\s\d+s$', t6))
```

```python
# . Matches any character
t7 = 'Just check out this link: www.amazingpics.com. It has great photos.'
re.findall(r'www.+com', t7)
```

```python
# Capturing Groups : in ()
sen = 'Clary has 2 friends who she spend a lot time with. Susan has 3 brothers while John has 4 sisters.'
re.findall('([A-Za-z]+)\s\w+\s(\d+)\s(\w+)', sen)
```

```python
# apply quantifier to the group
re.search(r'(\d[A-Za-z])+', 'My user name is 3e4r4t5yt') # immediately to the left
```

```python
my_numbers = 'My numbers are 3453 and 88'
print(re.findall(r'(\d+)', my_numbers))
print(re.findall(r'(\d)', my_numbers))
print(re.findall(r'(\d)+', my_numbers))
```

```python
# OR |
t8 = 'I want 2 dog, 3 bird and a cat'
print(re.findall(r'dog|bird|cat', t8))
print(re.findall(r'\d+\sdog|bird|cat', t8)) # why : because OR separate all text include \d+\s
print(re.findall(r'\d*\sdog|\d*\sbird|\d*\scat', t8))
print(re.findall(r'(\d*)\s(bird|dog|cat)', t8))
```

```python
# Non capturing : ?:
t9 = 'John Smith: 34-34-34-042-980, Rebeca Smith: 10-10-10-434-425'
print(re.findall(r'(?:\d{2}-){3}(\d{3}-\d{3})', t9))

t10 = 'To day is 23rd May 2019. Tomorrow is 24th May 19'
print(re.findall(r'(\d+)(?:th|rd)', t10))
```
